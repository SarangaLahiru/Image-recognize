```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Umang-19/devjam/main/public/mldata.csv', encoding='latin-1')
```

```python
data.head()
```

```python
columns_to_dummify = ['self-learning capability?', 'Extra-courses did','certifications','workshops','reading and writing skills','interested career area ','memory capability score','Interested subjects','Type of company want to settle in?','Taken inputs from seniors or elders','Interested Type of Books','Management or Technical','hard/smart worker','worked in teams ever?','Introvert']
data = pd.get_dummies(data, columns=columns_to_dummify)
data = data.replace({True:1 , False:0})
data.head()
```

```python
data.shape
```

```python
X = data.drop(['Suggested Job Role'], axis=1)
y = data['Suggested Job Role']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(model+ ' --> ' +acc)
```

```python


from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()

data.head()
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_

best_model.score(X_test, y_test)
```
